### Load the cleaned data for each countries

```python
import pandas as pd

# Load cleaned datasets
benin_df = pd.read_csv("../data/benin_clean.csv", index_col="Timestamp", parse_dates=True)
sierra_df = pd.read_csv("../data/sierrleon_clean.csv", index_col="Timestamp", parse_dates=True)
togo_df = pd.read_csv("../data/togo_cleaned.csv", index_col="Timestamp", parse_dates=True)

# Display first few rows of each
display("Benin Data:", benin_df.head())
display("Sierra Leone Data:", sierra_df.head())
display("Togo Data:", togo_df.head())
```

### Metric Comparison

##### Boxplots for GHI, DNI, DH

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Combine data into a single DataFrame for plotting
data = pd.concat([benin_df["GHI"].rename("Benin"), sierra_df["GHI"].rename("Sierra Leone"), togo_df["GHI"].rename("Togo")], axis=1)
plt.figure(figsize=(10, 6))
sns.boxplot(data=data)
plt.title("Boxplot of GHI by Country")
plt.ylabel("GHI (W/m²)")
plt.show()

# Repeat for DNI and DHI
data_dni = pd.concat([benin_df["DNI"].rename("Benin"), sierra_df["DNI"].rename("Sierra Leone"), togo_df["DNI"].rename("Togo")], axis=1)
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_dni)
plt.title("Boxplot of DNI by Country")
plt.ylabel("DNI (W/m²)")
plt.show()

data_dhi = pd.concat([benin_df["DHI"].rename("Benin"), sierra_df["DHI"].rename("Sierra Leone"), togo_df["DHI"].rename("Togo")], axis=1)
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_dhi)
plt.title("Boxplot of DHI by Country")
plt.ylabel("DHI (W/m²)")
plt.show()
```

### Summary Table:

```python
# Calculate summary statistics
stats = pd.DataFrame({
    "Benin": [benin_df["GHI"].mean(), benin_df["GHI"].median(), benin_df["GHI"].std()],
    "Sierra Leone": [sierra_df["GHI"].mean(), sierra_df["GHI"].median(), sierra_df["GHI"].std()],
    "Togo": [togo_df["GHI"].mean(), togo_df["GHI"].median(), togo_df["GHI"].std()]
}, index=["Mean", "Median", "Std Dev"])

# Extend to DNI and DHI
stats_dni = pd.DataFrame({
    "Benin": [benin_df["DNI"].mean(), benin_df["DNI"].median(), benin_df["DNI"].std()],
    "Sierra Leone": [sierra_df["DNI"].mean(), sierra_df["DNI"].median(), sierra_df["DNI"].std()],
    "Togo": [togo_df["DNI"].mean(), togo_df["DNI"].median(), togo_df["DNI"].std()]
}, index=["Mean", "Median", "Std Dev"])

stats_dhi = pd.DataFrame({
    "Benin": [benin_df["DHI"].mean(), benin_df["DHI"].median(), benin_df["DHI"].std()],
    "Sierra Leone": [sierra_df["DHI"].mean(), sierra_df["DHI"].median(), sierra_df["DHI"].std()],
    "Togo": [togo_df["DHI"].mean(), togo_df["DHI"].median(), togo_df["DHI"].std()]
}, index=["Mean", "Median", "Std Dev"])

# Display tables
print("GHI Summary Statistics:")
display(stats)
print("DNI Summary Statistics:")
display(stats_dni)
print("DHI Summary Statistics:")
display(stats_dhi)
```

### Statistical Testing

#### One-Way ANOVA:

```python
from scipy.stats import f_oneway

# Perform ANOVA on GHI
f_stat, p_value = f_oneway(benin_df["GHI"].dropna(), sierra_df["GHI"].dropna(), togo_df["GHI"].dropna())
print(f"ANOVA Results for GHI: F-statistic = {f_stat:.4f}, p-value = {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("The differences in GHI across countries are statistically significant.")
else:
    print("The differences in GHI across countries are not statistically significant.")
```

### Key Observations

### Key Observations
* Togo exhibits the highest average GHI (454.08 W/m²) and DNI (296.41 W/m²), indicating strong solar potential, but its high variability (Std Dev 319.10 for GHI) and outliers suggest challenges in consistent energy production.
* Benin and Sierra Leone show similar GHI means (445.47 W/m²) and medians (406.10 W/m²), with Benin’s seasonal peaks (e.g., February-April) offering potential, though both have significant variability (Std Dev 294.38) due to outliers.
* Sierra Leone’s lower DHI (mean 251.60 W/m²) and presence of outliers may indicate inconsistent diffuse light, making it less reliable compared to Benin and Togo for diverse solar applications.

### Bar Chart Ranking by Average GHI

```python
# Calculate average GHI
avg_ghi = pd.Series({
    "Benin": benin_df["GHI"].mean(),
    "Sierra Leone": sierra_df["GHI"].mean(),
    "Togo": togo_df["GHI"].mean()
})

# Plot bar chart
plt.figure(figsize=(8, 5))
avg_ghi.sort_values(ascending=False).plot(kind="bar")
plt.title("Ranking of Countries by Average GHI")
plt.ylabel("Average GHI (W/m²)")
plt.xlabel("Country")
plt.show()
```
